## 0. ÌôòÍ≤Ω ÏÑ§Ï†ï

````python
# ÌïÑÏöî Ìå®ÌÇ§ÏßÄ ÏÑ§Ïπò (ÏµúÏ¥à 1Ìöå)
# !pip install xgboost lightgbm scikit-learn pandas numpy matplotlib seaborn

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns
import os

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import (
    accuracy_score, roc_auc_score, f1_score,
    classification_report, ConfusionMatrixDisplay, confusion_matrix
)
import xgboost as xgb
import lightgbm as lgb

# ÌïúÍµ≠Ïñ¥ Ìè∞Ìä∏
plt.rcParams['axes.unicode_minus'] = False
try:
    plt.rcParams['font.family'] = 'NanumGothic'
except:
    pass

print('‚úÖ ÌôòÍ≤Ω ÏÑ§Ï†ï ÏôÑÎ£å')
````

## 1. Îç∞Ïù¥ÌÑ∞ Î°úÎìú

> Ïã§Ï†ú ÏÇ¨Ïö© Ïãú: ÌïúÍµ≠ÎßàÏÇ¨Ìöå Í≥µÍ≥µÎç∞Ïù¥ÌÑ∞ CSVÎ•º `pd.read_csv()`Î°ú ÏùΩÏñ¥Ïò§ÏÑ∏Ïöî.
> ÏïÑÎûòÎäî ÎèôÏùºÌïú Íµ¨Ï°∞Ïùò **ÏÉòÌîå Îç∞Ïù¥ÌÑ∞**Î•º ÏÉùÏÑ±Ìï©ÎãàÎã§.

````python
# ‚îÄ‚îÄ Ïã§Ï†ú Îç∞Ïù¥ÌÑ∞ ÏÇ¨Ïö© Ïãú Ï£ºÏÑù Ìï¥Ï†ú ‚îÄ‚îÄ
# df = pd.read_csv('data/raw/kra_race_results.csv', encoding='utf-8-sig')

# ‚îÄ‚îÄ ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ± (Jupyter import Í≤ΩÎ°ú Í≥†Ï†ï) ‚îÄ‚îÄ
import sys
import os
ROOT = os.path.abspath(os.getcwd())
pred_dir = os.path.join(ROOT, 'src', 'prediction')
if not os.path.isdir(pred_dir):
    ROOT = os.path.dirname(ROOT)
    pred_dir = os.path.join(ROOT, 'src', 'prediction')
sys.path.insert(0, pred_dir)
from horse_racing_top3 import generate_kra_sample, engineer_features, FEATURE_COLS, TARGET

df_raw = generate_kra_sample(n_races=600, horses_per_race=10, seed=42)

print(f'Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞: {df_raw.shape}')
print(f'Í≤ΩÏ£º Ïàò   : {df_raw["race_id"].nunique():,}Í∞ú')
print(f'Top3 ÎπÑÏú® : {df_raw["is_top3"].mean():.1%}')
df_raw.head()
````

## 2. ÌÉêÏÉâÏ†Å Îç∞Ïù¥ÌÑ∞ Î∂ÑÏÑù (EDA)

````python
fig, axes = plt.subplots(2, 3, figsize=(16, 9))
fig.suptitle('ÏäπÎßà Í≤ΩÏ£º Îç∞Ïù¥ÌÑ∞ EDA', fontsize=16, fontweight='bold')

# 1) Í≤ΩÎßàÏû•Î≥Ñ Í≤ΩÏ£º Ïàò
track_counts = df_raw['track'].value_counts()
axes[0,0].bar(track_counts.index, track_counts.values, color=['#4472C4','#ED7D31','#A9D18E'])
axes[0,0].set_title('Í≤ΩÎßàÏû•Î≥Ñ Í≤ΩÏ£º Ïàò')
axes[0,0].set_ylabel('Í≤ΩÏ£º Ïàò')

# 2) Í±∞Î¶¨Î≥Ñ Top3 ÎπÑÏú®
top3_by_dist = df_raw.groupby('distance')['is_top3'].mean().reset_index()
axes[0,1].bar(top3_by_dist['distance'].astype(str), top3_by_dist['is_top3'], color='#4472C4')
axes[0,1].axhline(0.3, color='red', linestyle='--', label='ÌèâÍ∑†(30%)')
axes[0,1].set_title('Í±∞Î¶¨Î≥Ñ Top3 ÏßÑÏûÖ ÎπÑÏú®')
axes[0,1].set_ylabel('Top3 ÎπÑÏú®')
axes[0,1].legend()

# 3) Îßê ÎÇòÏù¥ Î∂ÑÌè¨
axes[0,2].hist(df_raw['age'], bins=range(2,10), color='#4472C4', edgecolor='white')
axes[0,2].set_title('Îßê ÎÇòÏù¥ Î∂ÑÌè¨')
axes[0,2].set_xlabel('ÎÇòÏù¥ (ÏÑ∏)')
axes[0,2].set_ylabel('ÎπàÎèÑ')

# 4) Í∏∞Ïàò ÏäπÎ•† vs Top3 Ïó¨Î∂Ä
top3_yes = df_raw[df_raw['is_top3']==1]['jockey_win_rate']
top3_no  = df_raw[df_raw['is_top3']==0]['jockey_win_rate']
axes[1,0].hist(top3_yes, alpha=0.6, label='Top3', color='#4472C4', bins=20)
axes[1,0].hist(top3_no,  alpha=0.6, label='Non-Top3', color='#ED7D31', bins=20)
axes[1,0].set_title('Í∏∞Ïàò ÏäπÎ•† Î∂ÑÌè¨ (Top3 Ïó¨Î∂Ä)')
axes[1,0].set_xlabel('Í∏∞Ïàò ÏäπÎ•†')
axes[1,0].legend()

# 5) Î∞∞ÎãπÎ•† vs Top3
axes[1,1].boxplot(
    [df_raw[df_raw['is_top3']==1]['win_odds'].clip(0,30),
     df_raw[df_raw['is_top3']==0]['win_odds'].clip(0,30)],
    labels=['Top3', 'Non-Top3'],
    patch_artist=True,
    boxprops=dict(facecolor='#DEEAF1')
)
axes[1,1].set_title('Î∞∞ÎãπÎ•† Î∂ÑÌè¨ (Top3 vs Non-Top3)')
axes[1,1].set_ylabel('Îã®Ïäπ Î∞∞ÎãπÎ•†')

# 6) Ï£ºÎ°úÎ≥Ñ Top3 ÎπÑÏú®
top3_surface = df_raw.groupby('surface')['is_top3'].mean()
axes[1,2].bar(top3_surface.index, top3_surface.values, color=['#4472C4','#ED7D31'])
axes[1,2].set_title('Ï£ºÎ°úÎ≥Ñ Top3 ÎπÑÏú®')
axes[1,2].set_ylabel('Top3 ÎπÑÏú®')

plt.tight_layout()
plt.savefig('../outputs/eda_charts.png', dpi=150, bbox_inches='tight')
plt.show()
print('‚úÖ EDA ÏãúÍ∞ÅÌôî ÏôÑÎ£å')
````

## 3. ÌîºÏ≤ò ÏóîÏßÄÎãàÏñ¥ÎßÅ

````python
df = engineer_features(df_raw)

X = df[FEATURE_COLS]
y = df[TARGET]

print('ÌîºÏ≤ò Î™©Î°ù:')
for i, col in enumerate(FEATURE_COLS, 1):
    print(f'  {i:2d}. {col}')
print(f'\nÏ¥ù ÌîºÏ≤ò Ïàò: {len(FEATURE_COLS)}Í∞ú')
print(f'ÌïôÏäµ ÏÉòÌîå: {len(X):,}Í∞ú')
````

## 4. Î™®Îç∏ ÌïôÏäµ

````python
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(f'ÌïôÏäµ ÏÑ∏Ìä∏: {len(X_tr):,}Ìñâ  |  ÌÖåÏä§Ìä∏ ÏÑ∏Ìä∏: {len(X_te):,}Ìñâ')

# XGBoost
print('\n[XGBoost ÌïôÏäµ Ï§ë...]')
pos_w = (y_tr==0).sum() / max((y_tr==1).sum(), 1)
xgb_model = xgb.XGBClassifier(
    n_estimators=400, max_depth=5, learning_rate=0.04,
    subsample=0.8, colsample_bytree=0.75,
    scale_pos_weight=pos_w, eval_metric='logloss',
    verbosity=0, random_state=42
)
xgb_model.fit(X_tr, y_tr)
print('  XGBoost ÏôÑÎ£å')

# LightGBM
print('[LightGBM ÌïôÏäµ Ï§ë...]')
lgb_model = lgb.LGBMClassifier(
    n_estimators=400, max_depth=5, learning_rate=0.04,
    subsample=0.8, colsample_bytree=0.75,
    class_weight='balanced', verbose=-1, random_state=42
)
lgb_model.fit(X_tr, y_tr)
print('  LightGBM ÏôÑÎ£å')

print('\n‚úÖ Î™®Îç∏ ÌïôÏäµ ÏôÑÎ£å')
````

## 5. Î™®Îç∏ ÌèâÍ∞Ä

````python
# ÏïôÏÉÅÎ∏î ÏòàÏ∏°
y_proba = np.mean([
    xgb_model.predict_proba(X_te)[:, 1],
    lgb_model.predict_proba(X_te)[:, 1]
], axis=0)
y_pred = (y_proba >= 0.5).astype(int)

acc = accuracy_score(y_te, y_pred)
f1  = f1_score(y_te, y_pred, zero_division=0)
auc = roc_auc_score(y_te, y_proba)

print('=' * 40)
print(f'  Accuracy : {acc:.4f}')
print(f'  F1 Score : {f1:.4f}')
print(f'  ROC-AUC  : {auc:.4f}')
print('=' * 40)
print()
print(classification_report(y_te, y_pred, target_names=['Non-Top3','Top3'], zero_division=0))

# ÌòºÎèô ÌñâÎ†¨
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ConfusionMatrixDisplay(confusion_matrix(y_te, y_pred),
                       display_labels=['Non-Top3','Top3']).plot(
    ax=axes[0], colorbar=False, cmap='Blues'
)
axes[0].set_title('Confusion Matrix', fontsize=13, fontweight='bold')

# ÌîºÏ≤ò Ï§ëÏöîÎèÑ
fi = pd.DataFrame({'feature': FEATURE_COLS,
                   'importance': xgb_model.feature_importances_}
                 ).sort_values('importance', ascending=False).head(15)
axes[1].barh(fi['feature'][::-1], fi['importance'][::-1], color='#4472C4')
axes[1].set_title('ÌîºÏ≤ò Ï§ëÏöîÎèÑ Top 15', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.savefig('../outputs/model_eval.png', dpi=150, bbox_inches='tight')
plt.show()
````

## 6. ÍµêÏ∞® Í≤ÄÏ¶ù (5-Fold)

````python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

xgb_cv = cross_val_score(xgb_model, X, y, cv=cv, scoring='roc_auc')
lgb_cv = cross_val_score(lgb_model, X, y, cv=cv, scoring='roc_auc')

print('5-Fold Cross Validation ROC-AUC:')
print(f'  XGBoost  : {xgb_cv.mean():.4f} ¬± {xgb_cv.std():.4f}')
print(f'  LightGBM : {lgb_cv.mean():.4f} ¬± {lgb_cv.std():.4f}')

# ÏãúÍ∞ÅÌôî
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(range(1,6), xgb_cv, 'o-', label='XGBoost', color='#4472C4')
ax.plot(range(1,6), lgb_cv, 's-', label='LightGBM', color='#ED7D31')
ax.axhline(xgb_cv.mean(), color='#4472C4', linestyle='--', alpha=0.5)
ax.axhline(lgb_cv.mean(), color='#ED7D31', linestyle='--', alpha=0.5)
ax.set_title('5-Fold CV ROC-AUC', fontsize=13, fontweight='bold')
ax.set_xlabel('Fold')
ax.set_ylabel('ROC-AUC')
ax.legend()
ax.set_ylim(0.5, 1.0)
plt.tight_layout()
plt.show()
````

## 7. Í≤ΩÏ£º Top 3 ÏòàÏ∏° (Ïã§Ï†ú ÏÇ¨Ïö© ÏòàÏãú)

````python
import sys, os
pred_dir = os.path.join(os.path.abspath(os.getcwd()), 'src', 'prediction')
if not os.path.isdir(pred_dir):
    pred_dir = os.path.join(os.path.dirname(os.getcwd()), 'src', 'prediction')
sys.path.insert(0, pred_dir)
from horse_racing_top3 import predict_top3_for_race

# ÏÉòÌîå Í≤ΩÏ£º ÏÑ†ÌÉù
sample_race_id = df['race_id'].iloc[0]
sample_race = df[df['race_id'] == sample_race_id].copy()

# ÏòàÏ∏°
result = predict_top3_for_race(sample_race, [xgb_model, lgb_model])

# Ïã§Ï†ú Í≤∞Í≥º Î≥ëÌï©
actual = df_raw[df_raw['race_id'] == sample_race_id][['horse_no','finish_pos']]
result = result.merge(actual, on='horse_no')
result['actual_top3'] = result['finish_pos'].apply(lambda x: '‚úÖ' if x <= 3 else '')

print(f'Í≤ΩÏ£º ID: {sample_race_id}')
print()
print(result[['horse_no','horse_name','win_odds','top3_prob','prob_rank','finish_pos','actual_top3','recommendation']].to_string(index=False))
````
